```python
import pandas as pd
import numpy as np

# Compute the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Create a mask for highly correlated features (absolute correlation > 0.9)
threshold = 0.9
high_corr_features = set()

# Iterate over the correlation matrix
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:  # Check if correlation is above threshold
            colname = correlation_matrix.columns[i]  # Get the feature name
            high_corr_features.add(colname)  # Add to the set of features to drop

# Drop highly correlated features
df_reduced = df.drop(columns=high_corr_features)

print(f"Dropped features: {high_corr_features}")
print(f"Remaining features: {df_reduced.columns.tolist()}")
```

```python
from sklearn.model_selection import train_test_split
#“diagnosis” column is your label, split your data into train test accordingly
input_features = df.drop('diagnosis', axis=1)  
target_labels = df['diagnosis']  
train_features, test_features, train_labels, test_labels = train_test_split(input_features, target_labels, test_size=0.2, random_state=42)

# Print the size of the training set and testing set
print("Size of Training Set:", train_features.shape)
print("Size of Testing Set:", test_features.shape)
```

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from matplotlib.colors import LinearSegmentedColormap

#Apply K-means algorithm on the entire preprocessed data
numeric_data = df.select_dtypes(include=[np.number]).fillna(df.select_dtypes(include=[np.number]).mean())
categorical_data = df.select_dtypes(exclude=[np.number]).fillna(df.select_dtypes(exclude=[np.number]).mode().iloc[0])
combined_data = pd.concat([numeric_data, categorical_data], axis=1)
scaled_data = StandardScaler().fit_transform(numeric_data)
pca_data = PCA(n_components=2, random_state=42).fit_transform(scaled_data)
kmeans_labels = KMeans(n_clusters=2, random_state=42).fit_predict(pca_data)
custom_cmap = LinearSegmentedColormap.from_list("green_maroon", ["green", "maroon"])


#plot a graph displaying clustering in different colors
plt.figure(figsize=(8, 6))
plt.scatter(pca_data[:, 0], pca_data[:, 1], c=kmeans_labels, cmap=custom_cmap, alpha=0.6)
plt.title('PCA and K-means Clustering')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.show()
```

```python
#Classification algorithms to get better results - random forest, logistic, SVC

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression

scaled_data_frame = pd.DataFrame(scaled_data, columns=numeric_data.columns)

merged_data = pd.concat([scaled_data_frame, categorical_data.reset_index(drop=True)], axis=1)
feature_inputs = merged_data.drop(columns=["diagnosis", "id"]) 
label_targets = merged_data["diagnosis"].map({"M": 1, "B": 0})  

train_set_features, test_set_features, train_set_labels, test_set_labels = train_test_split(
    feature_inputs, label_targets, test_size=0.3, random_state=42
)

scaler = StandardScaler() 
train_set_features_scaled = scaler.fit_transform(train_set_features)
test_set_features_scaled = scaler.transform(test_set_features)

# Here I am using 3 algorithms: Support Vector Machine (SVM), Random Forest, and Logistic Regression

classification_models = {
    "Support Vector Machine": SVC(kernel='linear', C=1.0),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42),
    "Logistic Regression": LogisticRegression(solver='liblinear', C=0.5)
}
# Initialize an empty dictionary to store the accuracy of each model
model_accuracies = {}
for model_identifier, model_instance in classification_models.items():
    model_features = train_set_features_scaled if model_identifier != "Random Forest" else train_set_features
    model_test_features = test_set_features_scaled if model_identifier != "Random Forest" else test_set_features
    
    model_instance.fit(model_features, train_set_labels)
    model_predictions = model_instance.predict(model_test_features)
    model_accuracy = accuracy_score(test_set_labels, model_predictions)
    model_accuracies[model_identifier] = model_accuracy
    print(f"Accuracy of the {model_identifier}: {model_accuracy:.2f}")
```

```python
#Plot a graph comparing accuracies of the above three models - Support Vector Machine (SVM), Random Forest, and Logistic Regression
plt.figure(figsize=(8, 5))
plt.bar(model_accuracies.keys(), model_accuracies.values(), color=['Navy', 'Purple','Green'])
plt.ylim(0, 1)
plt.ylabel('Accuracy')
plt.title('Model Accuracies')
plt.grid(axis='y')
plt.show()
```

```python
#choosing label as 'compactness_mean'
target_col = 'compactness_mean'
features_df = df.drop(columns=[target_col, 'id', 'diagnosis'])  
target_series = df[target_col]  
features_train, features_test, target_train, target_test = train_test_split(features_df, target_series, test_size=0.2, random_state=42)
print("Training set size:", features_train.shape)
print("Testing set size:", features_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import mean_squared_error

train_feat, test_feat, train_targ, test_targ = train_test_split(features_df, target_series, test_size=0.2, random_state=42)

## Here I am using 2 algorithms: Gradient Boosting and Random Forest
models_dict = {
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42)
}

param_grid_rf = {
    'n_estimators': [50, 100],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5],
}
param_grid_gb = {
    'n_estimators': [50, 100],
    'learning_rate': [0.01, 0.1, 0.2],
    'max_depth': [3, 5, 7],
}

#Dict to store mean square losses for 2 algorithms mentioned above
best_models_dict = {}

grid_rf = GridSearchCV(RandomForestRegressor(random_state=42), param_grid_rf, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_rf.fit(train_feat, train_targ)
best_models_dict["Random Forest"] = grid_rf.best_estimator_
grid_gb = GridSearchCV(GradientBoostingRegressor(random_state=42), param_grid_gb, cv=5, scoring='neg_mean_squared_error', n_jobs=-1)
grid_gb.fit(train_feat, train_targ)
best_models_dict["Gradient Boosting"] = grid_gb.best_estimator_

print("All models have been successfully trained and are stored in best_models_dict.")
```

```python
from sklearn.metrics import mean_squared_error
mse_dict = {}

#Printing mean squared loss for algorithms - Gradient Boosting, Random Forest, and Logistic Regression
for model_name, model in best_models_dict.items():
    preds = model.predict(test_feat)
    mse = mean_squared_error(test_targ, preds)
    mse_dict[model_name] = mse
    print(f"MSE for {model_name}: {mse}")
```

```python
models = list(mse_dict.keys())
mse_values_list = list(mse_dict.values())

plt.figure(figsize=(10, 6))
bars = plt.bar(models, mse_values_list, color=['Brown', 'teal', 'Olive'], width=0.6)
plt.title('Comparison of Regression Models Based on Mean Squared Error')
plt.xlabel('Regression Models')
plt.ylabel('Mean Squared Error')
plt.ylim(0, max(mse_values_list) + 1e-4)  
plt.grid(axis='y', linestyle='--', alpha=0.7)

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 1e-5, round(yval, 10), ha='center', va='bottom', fontsize=12)

plt.show()
```
